```python
df = df.drop(columns=["Unnamed: 0"])
```

```python
# removing the values of bidder " "
df_new = df[df['bidder'] != " "].reset_index(drop=True)
```

```python
# filling the prebid win count with zero
df_new['prebid_win_count'] = df_new['prebid_win_count'].fillna(0)
df_new['win_count'] = df_new['win_count'].fillna(0)
```

```python
df_new['bidder'].unique()
```

```python
data =  df_new[["bidder", "request_count", "response_count","prebid_win_count","win_count"]]
```

```python
data.isna().sum()
```

```python
final_df = data.groupby("bidder").agg({"request_count":"sum" ,
                            "prebid_win_count" :"sum",
                            "response_count":"sum",
                            "prebid_win_count":"sum",
                            "win_count":"sum"}).reset_index()
```

```python
final_df
```

```python
# calculating the win rate in final_Df
final_df['win_rate'] = final_df['win_count']*100/ final_df['request_count']
final_df['win_rate'] = final_df['win_rate'].round(2)

final_df['prebid_win_rate'] = final_df['prebid_win_count']*100/ final_df['request_count']
final_df['prebid_win_rate'] = final_df['prebid_win_rate'].round(2)
```

```python
final_df.sort_values(by=["win_rate","prebid_win_rate"], ascending=[False,False]).reset_index(drop=True)
```

```python
result_df  = final_df[['bidder','prebid_win_rate','win_rate']].sort_values(by="prebid_win_rate", ascending=False).reset_index(drop=True)
result_df.head(2)
```
